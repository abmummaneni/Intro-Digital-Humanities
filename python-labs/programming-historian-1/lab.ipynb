```python
# sample search (should yield two results)
query = "peter parley date:[1825 TO 1830] mediatype:texts"
results = ia.search_items(query)
vol_ids = []
for result in results:
        vol_ids.append(result['identifier'])
print(vol_ids)
item_id = vol_ids[2]
```

```python
returned_files = ia.get_files(vol_ids[2], formats=["Abbyy GZ"])
file = list(returned_files)[0]
file
```

```python
import os
ia.download(item_id, formats=["Abbyy GZ"], ignore_existing=True, destdir=os.getcwd(), no_directory=True)
```

```python
# pip install lxml
from lxml import etree
import gzip
img_pages = []
with gzip.open(file.name, 'rb') as fp:
        tree = etree.parse(fp)
        document = tree.getroot()
        for i, page in enumerate(document):
                for block in page:
                        if block.get("blockType") == "Picture":
                                img_pages.append(i)
                

```
